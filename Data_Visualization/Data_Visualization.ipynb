```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# 1 Matplotlib Functional Method
```

```python
plt.plot(x,y,'r--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
```

```python
#Multi Plots
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
# 2 Matplotlib Object Oriented Method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y,'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
#Insert Plotting
```

```python
fig = plt.figure()

axes1 =  fig.add_axes([0.1,0.1,0.8,0.8]) #Main Axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #Insert Axes

# Larger Figure Axes 1
axes.plot(x,y,'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_Label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('X_label_Axes2')
axes2.set_ylabel('Y_able_axes2')
axes2.set_title('Axes 2 Title')
```

```python
#Subplot()
```

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
x = np.linspace(0,5,11)
y = x**2

fig, axes = plt.subplots(nrows = 1, ncols=2)

axes[0].plot(x,y,'b')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'r')
axes[1].set_title('Second Plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y)

for ax in axes:
    ax.plot(x,y,'g')
    ax.set_title('Title')
    ax.set_xlabel('X - axis')
    ax.set_xlabel('Y - axis')
```

```python
#Color
# 'b' : blue
# 'r' : red
# 'g' : green

# 'b.-' : blue dot
# 'g--' : green dashed line

fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**4, 'g--')
```

```python
#RGB HEX CODE
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color = 'blue', alpha = 0.5) # 50% opacity
ax.plot(x, x+2, color = '#8B008B')
ax.plot(x, x+3, color = '#FF8C00')
```

```python
#Line Style, Dash, Marker
```

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.plot(x, x+1, color = 'red', linewidth = 0.25)
ax.plot(x, x+2, color = 'red', linewidth = 0.50)
ax.plot(x, x+3, color = 'red', linewidth = 0.75)
ax.plot(x, x+4, color = 'red', linewidth = 1.50)
ax.plot(x, x+5, color = 'red', linewidth = 2.00)
```

```python
fig, ax = plt.subplots(figsize = (12,6))

ax.plot(x, x+1, color = 'red', linewidth = 3, linestyle = '--')
ax.plot(x, x+2, color = 'red', lw = 3, ls = '-.')
ax.plot(x, x+3, color = 'red', lw = 3, ls = ':')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

#custom dash
line, = ax.plot(x, x+8, color = 'black', lw = 1.5)

line.set_dashes([5,10,15,10]) #format: line lengths, space length,...
```

```python
fig, ax = plt.subplots(figsize = (12,6))

#Marker symbols: '+', 'o' , '*', 's', ',' , '.' , '1' , '2' , '3' , '4' , ....
ax.plot(x, x+9, color = 'blue', lw = 1, ls = '-', marker = '+', ms = 15, alpha = 0.8)
ax.plot(x, x+10, color = 'green', lw = 1.25, ls = '--', marker = 'o', ms = 5, alpha = 0.5)
ax.plot(x, x+11,color = 'red',lw= 1.25, ls = '--', marker = 's')
ax.plot(x, x+12,color ='green', lw = 1.25, ls = '--', marker = '1')
```

```python
fig, ax = plt.subplots(figsize = (12,6))

#Marker symbols: '+', 'o' , '*', 's', ',' , '.' , '1' , '2' , '3' , '4' , ....
ax.plot(x, x+9, color = 'blue', lw = 1, ls = '-', marker = '+', ms = 15, alpha = 0.8)
ax.plot(x, x+10, color = 'green', lw = 1.25, ls = '--', marker = 'o', ms = 5, alpha = 0.5)
ax.plot(x, x+11,color = 'red',lw= 1.25, ls = '--', marker = 's', markerfacecolor = 'blue')
ax.plot(x, x+12,color ='green', lw = 1.25, ls = '--', marker = '1', markeredgewidth = 5, ms = 30)
```

```python
#Plot range

fig, axes = plt.subplots(1,3, figsize = (12,4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('Default axes ranges')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('right axes')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0,100])
axes[2].set_xlim([2,5])
axes[2].set_title('Custom axes range')
```
